```python
#1

class Shape:
    def area(self):
        pass
    def perimeter(self):
        pass
    def display_info(self):
        pass
    def get_name(self):
        pass

class Circle(Shape): 
    def __init__(self, radius):
        if radius <= 0:
            raise ValueError('Ошибка - отрицательное значение')
        self.radius = radius
        self.pi = 3.14
    def area(self):
        return self.pi * self.radius ** 2
    def perimeter(self):
        return 2 * self.pi * self.radius
    def display_info(self):
        print(f"Circle: area = {self.area()}, perimeter = {self.perimeter()}")
class Rectangle(Shape): 
    def __init__(self, width, height):
        if width <= 0 or height <= 0:
            raise ValueError('Ошибка - отрицательное значение')
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(self):
        return 2 * (self.width + self.height)
    def display_info(self):
        print(f"Rectangle: area = {self.area()}, perimeter = {self.perimeter()}")

class Triangle(Shape):
    def __init__(self, a, b, c):
        if a <= 0 or b <= 0 or c <= 0:
            raise ValueError('Ошибка - отрицательное значение')
        if not (a + b > c and a + c > b and b + c > a):
            raise ValueError('Ошибка построение треугольника')
        self.a = a
        self.b = b
        self.c = c
    def area(self):
        p = self.perimeter() / 2
        return (p * (p - self.a) * (p - self.b) * (p - self.c)) ** 0.5
    def perimeter(self):
        return self.a + self.b + self.c
    def display_info(self):
        print(f"Triangle: area = {self.area()}, perimeter = {self.perimeter()}")

class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)
    
```

```python
ci1 = Circle(radius = 3)
print('area Circle', ci1.area())
print('perimeter Circle', ci1.perimeter())
print('------------')
kv1 = Rectangle(width = 2, height = 4)
print('area Rectangle', kv1.area())
print('perimeter Rectangle', kv1.perimeter())
print('------------')
tr1 = Triangle(a =4,b=7, c = 10)
print('area Triangle', tr1.area())
print('perimeter Triangle', tr1.perimeter())
print('------------')
sq1 = Square(side = 8)
print('area Square', sq1.area())
print('perimeter Square', sq1.perimeter())
print('------------')


circle = Circle(3)
rectangle = Rectangle(2, 4)
triangle = Triangle(4, 7, 10)
square = Square(8)

shapes = [circle, rectangle, triangle, square]

for shape in shapes:
    shape.display_info()
```

```python
#2 

class Employee:

    def __init__(self, id, name, position, hire_date, department):
        self.id = id
        self.name = name
        self.position = position
        self.hire_date = hire_date
        self.vacation_days = 28  
        self.department = department
    def calculate_salary(self):
        pass
    def has_medical_insurance(self):
        pass
    def calculate_vacation_pay(self):
        salary = self.calculate_salary()
        return round((salary / 30) * self.vacation_days, 2)
    def calculate_tax(self):
        salary = self.calculate_salary()
        return round(salary * 0.13, 2)

class FullTimeEmployee(Employee):
    def __init__(self, id, name, position, hire_date, department, monthly_salary, bonus, medical_insurance):
        super().__init__(id, name, position, hire_date, department)
        self.monthly_salary = monthly_salary
        self.bonus = bonus
        self.medical_insurance = True
    def calculate_salary(self):
        return round(self.monthly_salary + (self.bonus / 12),2)
    def calculate_medical_insurance(self):
        ins = 'Страхова есть'
        return ins
    def get_employee_info(self):
        print(f"FullTimeEmployee: \nОТДЕЛ: {self.department.name}, {self.position} {self.name} ID: {self.id}, дата найма: {self.hire_date} \nбонус =  {self.bonus}, з/п в месяц  =  {self.monthly_salary},  з/п средняя в месяц  = {self.calculate_salary()}, \nналоги в месяц =  {self.calculate_tax()}, отпускные (больничные) за год = {self.calculate_vacation_pay()} \n{self.calculate_medical_insurance()}")
            
class PartTimeEmployee(Employee):
    def __init__(self, id, name, position, hire_date, department, salary_per_hours, hours, day_work_per_month, medical_insurance):
        super().__init__(id, name, position, hire_date, department)
        self.salary_per_hours = salary_per_hours
        self.hours = hours
        self.medical_insurance = True
        self.day_work_per_month = day_work_per_month
    def calculate_salary(self):
        return round(self.salary_per_hours * self.hours * self.day_work_per_month,2)
    def calculate_medical_insurance(self):
        ins = 'Ограниченный соцпакет'
        return ins
    def get_employee_info(self):
        print(f"PartTimeEmployee: \nОТДЕЛ: {self.department.name}, {self.position} {self.name} ID: {self.id}, дата найма: {self.hire_date} \nsalary_per_hours = {self.salary_per_hours}, hours = {self.hours}, day_work_per_month = {self.day_work_per_month}\nз/п средняя в месяц  = {self.calculate_salary()}, \nналоги в месяц =  {self.calculate_tax()}, отпускные (больничные) за год = {self.calculate_vacation_pay()} \n{self.calculate_medical_insurance()}")

class CommissionEmployee (Employee):
    def __init__(self, id, name, position, hire_date, department, monthly_salary, plan,  medical_insurance):
        super().__init__(id, name, position, hire_date, department)
        self.monthly_salary = monthly_salary
        self.plan = plan
        self.medical_insurance = True
    def calculate_salary(self):
        salary = 0
        if self.plan < 0.8:
            salary == 0
        if self.plan > 0.8 and self.plan  < 1:
            salary = 0.2 * self.monthly_salary * self.plan
        if self.plan > 1:
            salary = 0.4* self.monthly_salary * self.plan
        return round(self.monthly_salary + salary,2)
    def calculate_medical_insurance(self):
        ins = 'Соцпакета нет'
        return ins
    def get_employee_info(self):
        print(f"CommissionEmployee: \nОТДЕЛ: {self.department.name}, {self.position} {self.name} ID: {self.id}, дата найма: {self.hire_date} \n Ставка = {self.monthly_salary},выполнение плана = {self.plan}\nз/п средняя в месяц  = {self.calculate_salary()}, \nналоги в месяц =  {self.calculate_tax()}, отпускные (больничные) за год = {self.calculate_vacation_pay()} \n{self.calculate_medical_insurance()}")
    
    
class Manager(FullTimeEmployee):
    def __init__(self, id, name, position, hire_date, department, monthly_salary, bonus, medical_insurance, management_bonus, training_budget):
        super().__init__(id, name, position, hire_date, department, monthly_salary, bonus, medical_insurance)
        self.management_bonus = management_bonus
        self.training_budget = training_budget
    def calculate_salary(self):
        return round(self.monthly_salary + (self.bonus / 12) + (self.management_bonus) / 12 ,2)
    def calculate_budget(self):
        tr = self.training_budget
        return round(tr,2)
    def get_employee_info(self):
        print(f"FullTimeEmployee: \nОТДЕЛ: {self.department.name}, {self.position} {self.name} ID: {self.id}, дата найма: {self.hire_date} \nбонус =  {self.bonus}, бонус руководства =  {self.management_bonus}, з/п в месяц  =  {self.monthly_salary},  з/п средняя в месяц  = {self.calculate_salary()}, \nналоги в месяц =  {self.calculate_tax()}, отпускные (больничные) за год = {self.calculate_vacation_pay()} \n{self.calculate_medical_insurance()}\nдоступный бюджет на обучение = {self.calculate_budget()}")


class Department:
    def __init__(self, name):
        self.name = name
        self.employees = []
    def add_employee(self, employee):
        self.employees.append(employee)
    def get_department_info(self):
        total_salary = 0
        for employee in self.employees:
            total_salary += employee.calculate_salary()
        average_salary = total_salary / len(self.employees)
        return f"Отдел: {self.name}, Сотрудников: {len(self.employees)}. Зарплатный бюджет {total_salary}. Средняя зарплата сотрудника: {average_salary}"
```
