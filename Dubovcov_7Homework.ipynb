```python
class Money:

    def __init__(self, rubles: float, kopeica: float):
        self.rubles = rubles 
        self.kopeica = kopeica 

    def balance(self):
        all_rubles = self.rubles + self.kopeica / 100
        #all_kopeica = self.kopeica % 100
        print(all_rubles)

    def slogenie(self, rubles, kopeica):
        self.rubles += rubles
        self.kopeica += kopeica
        return (self.rubles + self.kopeica / 100)

    def vichitanie(self, rubles, kopeica):
        self.rubles -= rubles
        self.kopeica -= kopeica
        if self.rubles + self.kopeica / 100 < 0:
            print('нет денег')
        else:
            return (self.rubles + self.kopeica / 100)

    def umnogenie(self, rubles, kopeica):
        self.rubles *= rubles
        self.kopeica *= kopeica
        return (self.rubles + self.kopeica / 100)

    def delenie(self, rubles, kopeica):
        x = self.rubles + self.kopeica / 100
        y = rubles + kopeica / 100
        return (x/y)

    def new_balance(self):
        if self.kopeica // 100 >= 1:
            self.rubles += self.kopeica // 100
        self.kopeica = self.kopeica % 100
        print(self.rubles, 'рублей,', self.kopeica, 'копеек')

    def umn_drob(self, x: float):
        y = self.rubles + self.kopeica / 100
        return(y * x)

    def del_drob(self, x: float):
        y = self.rubles + self.kopeica / 100
        return(y / x)

    def sravnenie(self, rubles, kopeica):
        x = self.rubles + self.kopeica / 100
        y = rubles + kopeica / 100        
        if self.kopeica // 100 >= 1:
            self.rubles += self.kopeica // 100
        self.kopeica = self.kopeica % 100
        
        if kopeica // 100 >= 1:
            rubles += kopeica // 100
        kopeica = kopeica % 100

        if x > y:
            print(self.rubles, 'рублей,', self.kopeica, 'копеек больше чем', rubles, 'рублей,', kopeica, 'копеек')
        else:
            print(rubles, 'рублей,', kopeica, 'копеек больше чем', self.rubles, 'рублей,', self.kopeica, 'копеек')
            
```

```python
# 2

class Matrix:

    def __init__(self, numbers: list):
        self.numbers = numbers 
        self.rows = len(numbers)  
        self.cols = len(numbers[0])

    def new_matrix(self):
        return self.numbers

    def slogenie(self,self2):
        
        if self.rows != self2.rows or self.cols != self2.cols:
            print('Ошибка, матрицы разного размера')
            return None
        result = []
        for i in range(self.rows):
            tmp = []
            for j in range(self.cols):
                tmp.append(self.numbers[i][j] + self2.numbers[i][j])
            result.append(tmp)       
        return Matrix(result)

    
    def umnogenie(self,chislo):
        
        result = []
        for i in range(self.rows):
            tmp = []
            for j in range(self.cols):
                tmp.append(self.numbers[i][j] * chislo)
            result.append(tmp)
        
        return Matrix(result)
```
