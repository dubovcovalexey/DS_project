```python
import sqlalchemy as db
```

```python
engine = db.create_engine('sqlite:///mydb2.db')
```

```python
conn = engine.connect()
```

```python
metadata = db.MetaData()
```

```python
dir(db)
```

```python
students = db.Table(
    'Students', metadata,
    db.Column('id', db.Integer, primary_key=True),
    db.Column('name', db.Text),
    db.Column('age', db.Integer),
    db.Column('average_point', db.Numeric)
)
students
```

```python
subjects = db.Table(
    'Subjects', metadata,
    db.Column('id', db.Integer, primary_key=True),
    db.Column('id_student', db.Integer),
    db.Column('name_subject', db.Text),
    db.Column('point', db.Integer)
)
subjects
```

```python
metadata.create_all(engine)
```
