```python
# модели
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LogisticRegression
from sklearn.svm import SVR, SVC
from sklearn.datasets import make_regression, make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score, f1_score, r2_score
```

```python
y = df_model['quality']
x = df_model.drop(columns = ['quality'])
```

```python
# РАЗБИВАЕМ ДАННЫЕ 80/20
x_reg_train, x_reg_test, y_reg_train, y_reg_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

x_clf_train, x_clf_test, y_clf_train, y_clf_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
print("МОДЕЛИ ЛИНЕЙНОЙ РЕГРЕССИИ")
print("-" * 60)

# Классическая линейная регрессия
print("1. Классическая линейная регрессия...")
model1 = LinearRegression()
model1.fit(x_reg_train, y_reg_train)
y_pred1 = model1.predict(x_reg_test)
mse1 = mean_squared_error(y_reg_test, y_pred1)
mae1 = mean_absolute_error(y_reg_test, y_pred1)
r2_1 = r2_score(y_reg_test, y_pred1)
reg_results.append(("Классическая линейная", model1, mse1, mae1, r2_1))
print(f"   MSE: {mse1:.2f}, MAE: {mae1:.2f}, R²: {r2_1:.3f}")
```

```python
# Ridge регрессия
print("2. Ridge регрессия (L2)...")
model2 = Ridge(alpha=1.0)
model2.fit(x_reg_train, y_reg_train)
y_pred2 = model2.predict(x_reg_test)
mse2 = mean_squared_error(y_reg_test, y_pred2)
mae2 = mean_absolute_error(y_reg_test, y_pred2)
r2_2 = r2_score(y_reg_test, y_pred2)
print(f"   MSE: {mse2:.2f}, MAE: {mae2:.2f}, R²: {r2_2:.3f}")
```

```python
# Lasso регрессия
print("3. Lasso регрессия (L1)...")
model3 = Lasso(alpha=0.1, max_iter=10000)
model3.fit(x_reg_train, y_reg_train)
y_pred3 = model3.predict(x_reg_test)
mse3 = mean_squared_error(y_reg_test, y_pred3)
mae3 = mean_absolute_error(y_reg_test, y_pred3)
r2_3 = r2_score(y_reg_test, y_pred3)
print(f"   MSE: {mse3:.2f}, MAE: {mae3:.2f}, R²: {r2_3:.3f}")
```

```python
#  SVR для регрессии 
print("4. SVM для регрессии (SVR)...")
model4 = SVR(kernel='linear', C=1.0)
model4.fit(x_reg_train, y_reg_train)
y_pred4 = model4.predict(x_reg_test)
mse4 = mean_squared_error(y_reg_test, y_pred4)
mae4 = mean_absolute_error(y_reg_test, y_pred4)
r2_4 = r2_score(y_reg_test, y_pred4)
print(f"   MSE: {mse4:.2f}, MAE: {mae4:.2f}, R²: {r2_4:.3f}")
```
