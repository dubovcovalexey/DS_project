```python
import random

class Array:
    def __init__(self, data):
        if type(data) ==  list:
            self.data = data
        elif type(data) ==  int:
            self.data = self.random(data)
        elif type(data) == tuple:
            x, y = data
            self.data = self.riad(x, y)
        else:
            print('ошибка')

    def random(self, num):
        arr = []
        for i in range(num):
            arr.append(random.randint(1, num))
        return arr

    def riad(self, x, y):
        arr = []
        for i in range(x, y+1):
            arr.append(i)
        return arr   



class Array_double (Array):
    def __init__(self, data):
        super().__init__(data)

    def new_matrix(self, array, rows):
        cols = len(array) // rows
        matrix = []
        for i in range(rows):
            start = i * cols
            end = start + cols
            row = array[start:end]
            matrix.append(row)
        return matrix    

    def summa(self):
        total = 0
        for row in self.data:
            for element in row:
                total += element
        return total

    def slogenie(self, num):
        for i in range(len(self.data)):
            for j in range(len(self.data[i])):
                self.data[i][j] += num
        return self.data

    def element_sum(self, matrix2):
        if len(self.data) != len(matrix2) or len(self.data[0]) != len(matrix2[0]):
            print("Ошибка: матрицы разного размера")
        total =[]
        for i in range(len(self.data)):
            new_total = []
            for j in range(len(self.data[i])):
                new_total.append(self.data[i][j] + matrix2[i][j])
            total.append(new_total)
        
        return total

        
    def element_umn(self, matrix2):
        if len(self.data) != len(matrix2) or len(self.data[0]) != len(matrix2[0]):
            print("Ошибка: матрицы разного размера")
        total =[]
        for i in range(len(self.data)):
            new_total = []
            for j in range(len(self.data[i])):
                new_total.append(self.data[i][j] * matrix2[i][j])
            total.append(new_total)
        
        return total    


    def beauty_matrix(self):
        for row in self.data:
            print(end="")
            for num in row:
                print(num, end=" ") 
            print("") 
```

```python
a = Array([1, 2, 3, 4, 5, 6, 7])
print(a.data)
b = Array(5)
#print(b.data)
print(b.data)
c = Array((1,15))
print(c.data)
print('--------------------')

a = Array_double([1, 2, 3, 4, 5, 6, 7])
print(a.new_matrix(a.data, 3))
print('--------------------')
b = Array_double(a.new_matrix(a.data, 3))
print('сумма элементов', b.summa())
print('--------------------')
print('сложение счислом', b.slogenie(5))
print('--------------------')
matrix2 = [[10, 11], [12, 13], [14, 15]]
c = Array_double(a.new_matrix(a.data, 3))
print('сложение матриц:\n', a.new_matrix(a.data, 3),'\n',matrix2,'\nОтвет:', c.element_sum(matrix2))
print('--------------------')
print('умножение матриц:\n', a.new_matrix(a.data, 3),'\n',matrix2,'\nОтвет:', c.element_umn(matrix2))
print('--------------------')
print('красивый вывод')
b.beauty_matrix()
```
