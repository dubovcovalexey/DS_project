```python
#Intenger/Float

a = 10.3
b = 3.2

# Преобразование
x =  int(a)
y =  int(b)
print(type(x))

print(f'{x} + {y} = {x + y}')  
print(f'{x} - {y} = {x - y}')  
print(f'{x} * {y} = {x * y}')  
print(f'{x} / {y} = {x / y}')  
print(f'{x} // {y} = {x // y}')  
print(f'{x} % {y} = {x % y}')  
print(f'{x} ** {y} = {x ** y}')  
print(f'{x} > {y} = {x > y}') 
print(f'{x} < {y} = {x < y}') 
print(f'модуль -10 = {abs(-10)}')
print(f'округление 3.5 =  {round(2.5)}') # в четную сторону
print(f'округление 2.555 =  {round(2.5,1)}') 
```

```python
#String
x = "hello world!"
print(x)
x = x.capitalize()
print('capitalize: первый символ в верхний регистр:',x)
x = 'hello world!'

x = x.casefold()
print('casefold: все в нижний регистр:',x)
x = 'hello world!'

x = x.center(30,'*')
print('center: центрирует строку добавляя символы:',x)
x = 'hello world!'

x = x.count("*")
print('count: считает количество указанных символов:',x)
x = 'hello world!'

x = x.encode()
print('encode: закодированная врсия строки:',x)
x = 'hello world!'

x = x.endswith('!')
print('endswith: последний символ:',x)
x = 'hello\tworld!'

x= x.expandtabs(2)
print('expandtabs: устанавивает размер табцлуции:',x)
x = 'hello world!'

x=x.find('w')
print('find: ищет символ:',x)
x = '{}{}'

x=x.format('hello ','world!')
print('format: форматирует строки, числа:',x)
x = 'hello world!'

new = {'name': 'MAX'}
x = 'hello world and {name}!'
print('format_map: форматирует значение из словаря:', x.format_map(new))
x = 'hello world!'

x = x.index('world')
print('index: индекс значения:', x)
x = 'hello world!'

print('isalnum: true - если все символы это буквы и цифры:',  x.isalnum(),
      '\nisalpha: true - если все символы это буквы',  x.isalpha(),
      '\nisascii: true - если все символы это ASCII',  x.isascii(),
      '\nisdecima: true - если все символы десятичные',  x.isdecimal(),
      '\nisdigit: true - если все символы это числа',  x.isdigit(),
      '\nisidentifier: true - если строка является идентификатором',  x.isidentifier(),
      '\nislower: true - если все символы строчные',  x.islower(),
      '\nisnumeric: true - если все символы цифры',  x.isnumeric(),
      '\nisprintable: true - если все символы печатные',  x.isprintable(),
      '\nisspace: true - если все символы пробелы',  x.isspace(),
      '\nistitle: true - каждое слово в строке с заглавной буквы',  x.istitle(),
      '\nisupper: true - если все символы заглавные',  x.isupper(),
      
     )
x = ['hello', 'world']


print('join: объединяет элементы списка в одну строку:', '**'.join(x))
x = 'hello world!'

print('ljust: левое выравнивание:', x.ljust(20))
x = 'Hello World!'

print('lower: переводит в нижний регистр:', x.lower())
x = '     hello world!'

print('lstrip: удаляет пробелы слева:', x.lstrip())
x = 'hello world!'

tr = str.maketrans('h', 'T')
print('maketrans: словарь для перевода translate:', x.translate(tr))

print('partition: разделяет строку на 3 части:', x.partition(' '))

print('replace: замена части строки над другую:', x.replace(' ', '*'))

print('rfind: индекс пследнего вхождения подстроки в строку:', x.rfind('l'))

print('rjust: выравнивание по правому краю:', x.rjust(20, '*'))

print('rpartition: разделение строки на 3 части с конца:', x.rpartition(' '))

print('rsplit: разделение строки с конца:', x.rsplit('l', 1))

print('rstrip: удаление пробелов справа:', '  hello  '.rstrip())

print('split: разделение строки на список:', x.split(' '))

x = 'hello\nworld!'
print('splitlines: разделение по строкам:', x.splitlines())
x = 'hello world!'

print('startswith: true - проверка начала строки:', x.startswith('hello'))

print('strip: удаление пробелов с обеих сторон:', '  hello  '.strip())
x = 'Hello World!'

print('swapcase: инверсия регистра:', x.swapcase())
x = 'hello world!'

print('title: заглавные буквы каждого слова:', x.title())

print('upper: преобразование в верхний регистр:', x.upper())

print('zfill: заполнение нулями слева:', x.zfill(20))

```

```python
#String
leagues = ['NFL', 'NBA', 'NHL', 'MLB']
print('исходные данные:', leagues)

print('append: добавление элемента в конец')
leagues.append('MLS')
print(leagues) 

print('clear: очистка всего списка:')
leagues_clear = leagues.copy()
leagues_clear.clear()
print(leagues_clear) 

print('count: подсчет количества элементов:', leagues.count('NBA'))

print('extend: расширение списка другим списком:')
leagues.extend(['NCAAF', 'NCAAB'])
print(leagues)

print('index: поиск индекса элемента:')
print(leagues.index('NHL'))  

print('insert: вставка элемента по индексу:')
leagues.insert(1, 'F1')
print(leagues) 

print('pop: удаление элемента:', leagues.pop(1))
print(leagues) 


print('remove: удаление первого найденного элемента:')
leagues.remove('NBA')
print(leagues) 

print('reverse: обратный порядок элементов:')
leagues.reverse()
print(leagues) 

print('sort: сортировка списка:')
leagues.sort()
print(leagues)  
```

```python
#Dictonary
sports = {'football': 'NFL', 'basketball': 'NBA', 'hockey': 'NHL', 'baseball': 'MLB'}
print('Исходный словарь:', sports)

print('clear: удаление всех элементов / copy: создание копии словаря:')
sports_copy = sports.copy()
sports_copy.clear()
print(sports_copy)  

print('fromkeys: создание словаря с ключами и общим значением:')
keys = ['football', 'basketball', 'hockey']
sports2 = dict.fromkeys(keys, '-')
print(sports2)  

print('get: получение значения по ключу:')
print(sports.get('basketball'))   
print(sports.get('volleyball'))    

print('items: получение пар ключ-значение:')
print(sports.items())  

print('keys: получение всех ключей:')
print(sports.keys())   

print('pop: удаление элемента по ключу:')
delete = sports.pop('hockey')
print('Удален:', delete) 
print(sports)  

print('popitem: удаление последней добавленной пары:')
last_delete = sports.popitem()
print('Удалена последняя пара:', last_delete)  
print(sports)  

print('setdefault: получение значения или установка по умолчанию:')
print(sports.setdefault('basketball'))  
print(sports.setdefault('soccer', 'MLS'))  
print(sports)  

print('update: обновление словаря новыми парами:')
sports.update({'college sport': 'NCAA', 'raсe': 'F1'})
print(sports) 

print('values: получение всех значений:')
print(sports.values()) 
```
