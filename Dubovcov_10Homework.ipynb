```python
class BankAccount:
    def __init__(self, currency, balance):
        self.currency = currency
        self.balance = balance

class PaymentSystem:
    def __init__(self, systemName: str):
        self.systemName = systemName

    def pay(self, summa):
        print(f'Платеж прошел успешно.{summa}')

    def tax_calculation(self, tax):
        pass
        
    def time_calculation(self, summa):
        pass

    def info(self):
        pass
        

class Visa(PaymentSystem):
    def __init__(self, currency: str, balance: int, summa: int):
        super().__init__("Visa")
        self.currency = currency
        self.balance = balance
        self.summa = summa
        
    def tax_calculation(self):
        if self.summa > 1000:
            tax = 0.05
        else:
            tax = 0.03
        tax_amount = tax * self.summa
        #print(f'комиссия: {tax_amount}')
        return tax_amount 

    def time_calculation(self):
         if self.currency == 'BYN':
             time = 6
         elif self.currency == 'USD':
             time = 14
         else:
             print('Ошибка платежа')
             return 0
         #print(f'{time} минут')
         return time

    def conversion(self):
        if self.currency == 'BYN':
            summa_new = self.summa
        elif self.currency == 'USD':
            summa_new = self.summa * 3
        else:
            print('Ошибка платежа')
            return 0
        conv = summa_new
        #print(f'Переведено {self.summa} {self.currency} = {summa_new} {self.currency}')
        return conv  

    def new_balance(self):
        new_balance = self.balance - self.conversion() - self.tax_calculation()
        return new_balance

    def info(self):
        print(f'Система: {self.systemName}')
        print(f'Баланс: {self.balance}')
        print(f'Время перевода: {self.time_calculation()} минут')
        print(f'Комиссия: {self.tax_calculation()}')
        print(f'Сумма конверсии: {self.conversion()}')
        print(f'Баланс после перевода: {self.new_balance()}')

class MC(PaymentSystem):
    def __init__(self, currency: str, balance: int, summa: int):
        super().__init__("MasterCard")
        self.currency = currency
        self.balance = balance
        self.summa = summa
        
    def tax_calculation(self):
        if self.summa > 1000:
            tax = 0.1
        else:
            tax = 0.03
        tax_amount = tax * self.summa
        #print(f'комиссия: {tax_amount}')
        return tax_amount 

    def time_calculation(self):
         if self.currency == 'BYN':
             time = 2
         elif self.currency == 'USD':
             time = 10
         else:
             print('Ошибка платежа')
             return 0
         #print(f'{time} минут')
         return time

    def conversion(self):
        if self.currency == 'BYN':
            summa_new = self.summa
        elif self.currency == 'USD':
            summa_new = self.summa * 3
        else:
            print('Ошибка платежа')
            return 0
        conv = summa_new
        #print(f'Переведено {self.summa} {self.currency} = {summa_new} {self.currency}')
        return conv  

    def new_balance(self):
        new_balance = self.balance - self.conversion() - self.tax_calculation()
        return new_balance
    
    def info(self):
        print(f'Система: {self.systemName}')
        print(f'Баланс: {self.balance}')
        print(f'Время перевода: {self.time_calculation()} минут')
        print(f'Комиссия: {self.tax_calculation()}')
        print(f'Сумма конверсии: {self.conversion()}')
        print(f'Баланс после перевода: {self.new_balance()}')
        
        

```
