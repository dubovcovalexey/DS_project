```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('medv', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
your_columns = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 
                'rad', 'tax', 'ptratio', 'b', 'lstat', 'medv']

for column in your_columns:
    plt.figure()  # это создает новое окно
    sns.histplot(data=df, x=column)
    plt.title(f'Распределение')
    
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.histplot(data=df, x="medv", ax=axes[0, 0])
sns.histplot(data=df, x="tax", ax=axes[1, 0])
sns.histplot(data=df, x="lstat", ax=axes[0, 1])
sns.histplot(data=df, x="b", ax=axes[1, 1])

```

```python
# Вычисляем средние значения
means = df[your_columns].mean()

# Создаем DataFrame для визуализации
means_df = means.reset_index()
means_df.columns = ['feature', 'mean_value']

# Строим график средних значений
plt.figure(figsize=(12, 6))
sns.barplot(data=means_df, x='feature', y='mean_value')
plt.title('Средние значения признаков Boston Housing')
plt.xticks(rotation=45)
plt.ylabel('Среднее значение')
plt.tight_layout()
plt.show()
```

```python
# Топ-5 наиболее коррелирующих с ценой признаков
corr_with_medv = df[your_columns].corr()['medv'].sort_values(ascending=False)
top_features = corr_with_medv.index[1:6] 

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, feature in enumerate(top_features):
    sns.scatterplot(data=df, x=feature, y='medv', ax=axes[i])
    axes[i].set_title(f'medv vs {feature}\n(corr: {corr_with_medv[feature]:.2f})')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='medv', kde=True)
plt.axvline(df['medv'].mean(), color='red', linestyle='--', label=f'Среднее: {df["medv"].mean():.2f}')
plt.title('Распределение цен на жилье (medv)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[your_columns])
plt.xticks(rotation=5)
plt.title('Распределение и выбросы признаков')
plt.tight_layout()
plt.show()
```

```python
key_features = ['crim', 'rm', 'age', 'dis', 'ptratio', 'lstat', 'medv']
sns.pairplot(df[key_features], diag_kind='hist')
plt.suptitle('Попарные отношения ключевых признаков', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['crim_level'] = pd.cut(df['crim'], bins=[0, 1, 5, 100], labels=['Низкий', 'Средний', 'Высокий'])
sns.scatterplot(data=df, x='rm', y='medv', hue='crim_level', palette='viridis')
plt.title('Зависимость цены от количества комнат и уровня преступности')
plt.legend(title='Уровень преступности')
plt.show()
```
