```python
matrices = [
    # 1
    np.array([[4, 0, 4, 5],
              [0, 1, 5, 6],
              [-2, 2, 2, 0],
              [1, -5, 1, 3]]),
    
    # 2
    np.array([[5, 0, 1, 3],
              [0, 1, 5, -1],
              [7, -1, 3, -2],
              [5, 3, 0, 1]]),
    
    # 3
    np.array([[4, 0, 1, 5],
              [1, 1, 0, 4],
              [0, 1, 1, 5],
              [6, -4, 1, 1]]),
    
    # 7
    np.array([[3, 2, 0, 1],
              [0, 2, 2, 3],
              [1, 5, 6, 1],
              [-2, 0, 3, 2]]),
    
    # 8
    np.array([[5, 1, 3, 0],
              [-3, 0, 5, 2],
              [2, 1, -1, 3],
              [1, 5, 0, 7]]),
    
    # 9
    np.array([[7, 6, 0, 1],
              [1, 3, 5, 0],
              [0, 1, -2, 4],
              [1, 1, 2, 3]]),
    
    # 13
    np.array([[2, 3, 0, 1],
              [-4, 0, 3, 1],
              [0, 8, 7, 5],
              [3, 3, 4, -2]]),
    
    # 14
    np.array([[3, 1, 3, 1],
              [2, 4, 2, 5],
              [1, 4, 5, 3],
              [-1, 0, 3, 2]]),
    
    # 15
    np.array([[7, 6, 0, 1],
              [1, 3, 5, 0],
              [0, 1, -2, 4],
              [1, 1, 2, 3]])
]

# Вычисляем определители
print("РЕЗУЛЬТАТЫ ВЫЧИСЛЕНИЯ ОПРЕДЕЛИТЕЛЕЙ:")
print("=" * 50)

for i, matrix in enumerate(matrices, 1):
    det = np.linalg.det(matrix)
    print(f"Матрица {i}:")
    print(f"Определитель = {det:.2f}")

    print("-" * 40)
```

```python
# 2

matrices = [
    # Вариант 1
    {
        'A': np.array([[1, 2, 4], [7, 5, -2], [-2, 1, 3]]),
        'B': np.array([[-1, 0, 3], [2, -6, 1], [-7, 1, 5]]),
        'C': np.array([[-2, 3], [4, 0], [7, -1]])
    },
    # Вариант 2
    {
        'A': np.array([[0, 3, 4], [-7, 1, -2], [2, -1, 3]]),
        'B': np.array([[1, 2, 4], [7, 0, -2], [-3, -2, 5]]),
        'C': np.array([[-3, 2], [0, 4], [-1, 7]])
    },
    # Вариант 3
    {
        'A': np.array([[5, 2, 9], [7, 0, -2], [-2, 1, 6]]),
        'B': np.array([[1, -2, 1], [7, 5, -2], [-2, 8, 3]]),
        'C': np.array([[-2, 3], [4, 5], [0, -1]])
    },
    # Вариант 4
    {
        'A': np.array([[1, -2, 8], [7, 5, -2], [-2, 3, 0]]),
        'B': np.array([[0, 2, 4], [1, -3, -2], [-2, 1, 3]]),
        'C': np.array([[0, 3], [5, -4], [7, -1]])
    },
    # Вариант 5
    {
        'A': np.array([[9, 2, 4], [7, 5, -5], [6, -2, 3]]),
        'B': np.array([[8, 2, 4], [0, 1, -3], [-2, 1, 6]]),
        'C': np.array([[-2, 0], [4, 8], [1, -1]])
    }
]

for i, matrices_set in enumerate(matrices, 1):
    print(f"\nВАРИАНТ {i}:")    
    A = matrices_set['A']
    B = matrices_set['B']
    C = matrices_set['C']
    
    # 1. A + B
    print("1. A + B =")
    print(A + B)
    
    # 2. B - A
    print("\n2. B - A =")
    print(B - A)
    
    # 3. A × C
    print("\n3. A × C =")
    print(np.dot(A, C))
    
    # 4. A × B × C
    print("\n4. A × B × C =")
    AB = np.dot(A, B)
    ABC = np.dot(AB, C)
    print(ABC)
    
    print("=" * 50)
```
