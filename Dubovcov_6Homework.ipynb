```python
# 4 

lst = [1, 2, 2, 3, 2, 1, 1]

def unique_in_order(lst: list):
    new_lst = []
    for q in range(len(lst)):
        if (q == 0 or lst[q] != lst[q-1]) and (q == len(lst)-1 or lst[q] != lst[q+1]) :
            new_lst.append(lst[q])
    return new_lst

unique_in_order(lst)
```

```python
# 5
ns: str = '1111 hello 321 world 3333 !22!'

def splitstr(ns):
    n = ns.split(' ')
    return n
    
ns2 = splitstr(ns)
 
def minlist(ns2):
    min_length = min(len(q) for q in ns2)
    new_lst = []
    for q in ns2:
        if len(q) == min_length:
            new_lst.append(q)
    return new_lst

minlist(ns2)
```

```python
#6 

chislo: int = 656


def root(chislo):
    if chislo < 10:
        return chislo
    total = 0   
    for i in "{}".format(chislo):
        total += int(i)    
    return root(total)

root(chislo)
```

```python
# 7

slovo: str = '12345'

def middle(slovo):
    lenght = len(slovo)
    index = lenght // 2
    if lenght % 2 == 0:
        return slovo[index - 1 : index + 1]
    else:  
        return slovo[index]

middle(slovo)
```

```python
# 8

prostoe_chislo: int = 4

def prosto(prostoe_chislo):
    if prostoe_chislo < 2:
        return False
    for i in range(2, prostoe_chislo):
        if prostoe_chislo % i == 0:
            return False
    return True
    
    

prosto(prostoe_chislo)
```
